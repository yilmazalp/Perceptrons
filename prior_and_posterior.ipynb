#### Probability mass function

We can use the probability mass function (PMF) of the Bernoulli distribution to get our desired probability for coin flips. The `probability mass function` is a function that gives the probabilities for `discrete random variables`. 

\begin{equation}
    P(X=x) = p^{x}(1-p)^{x}
\end{equation}

p takes values between 0 and 1. Because p represents probability values. x represents number of success. So, x denotes number of heads or tails in the event of flipping of a coin. 

It was the probability mass function for a `only one event`. So, we calculated the probability function for flipping of one coin. However, we want to find the probability of `all events` of flipping coin. So, if we interest n event then we calculate the probability of each event and multiply every event with another events since every event is independent from  each other. 

\begin{align}
    P(x_{1}, x_{2}, ... , x_{n} | \theta) &=\prod_{i}^{} P(x_{i}|\theta)\\
                                          &=\prod_{i}^{} \theta^{x_{i}} (1-\theta)^{1-x_{i}}
\end{align}

Now, we shall view expression of `st.bernoulli.pmf(data_coins_flips, p)`. In there, `data_coin_flips` is a sequence of data(coin is head or tail). It means that $x_{i}$ represents data point in `data_coin_flips` sequence. $\theta$ is $p$, means that probability of success. 

```python
# 0'dan baslayip 1 ile biten 100 adet sayi dizisi. cunku 'p' olasilik degerini temsil ediyor. yani 0 ve 1 arasinda deger alir. 
params = np.linspace(0, 1, 100) 
prod_prob=[]

#params dizisindeki her eleman hesaba katilarak butun para atma deneyindeki olasiliklarin carpimi hesaplanir
for p in params:
    prod_prob.append(np.product(st.bernoulli.pmf(data_coin_flips, p)))

#grafik
prod_prob = prod_prob/np.sum(prod_prob)
plt.plot(params, prod_prob)
sns.despine()
#plt.show()
```

Sampling distribution shows that distribution of available data. This data is observed values of flipped coins when coin is unfair because of p=0.8. However, prior distribution says that coin is fair since the distribution peaks at 0.5 . Now, we must find to posterior values and their distribution by Bayes theorem which we had explained it as before. 

Before go to posterior, we shall explain likelihood function. We know that likelihood function is that: 

$P(D|\theta)$: $Likelihood function$



$$P(D|\theta) = \theta^D(1-\theta)^{1-D}$$

$D$ : sample

$\theta$ : p

So, the code of `st.bernoulli.pmf(sample, p)` gives that `likelihood function` because sample is data which we have. If we interest the probabilites of all events then we use `np.product(st.bernoulli.pmf(sample, p))` which multiply the probabilites of all indepedent events.

                                    likelihood = np.array(np.product(st.bernoulli.pmf(sample, p)))

We created `sample` with binomial distribution which has probability value is 0.8. This is true probability value. 

#### Posterior 

We know that prior distribution values. So, we must know that likelihood of available data to calculate posterior because posterior is equal to multiplying prior and likelihood. 

\begin{equation}
    Posterior_{i} = \frac{prior_{i}*likelihood_{i}}{\text{normalization const.}}
\end{equation}

                                        posterior = [prior[i] * likelihood[i] for i in range(prior.shape[0])]
                                        posterior = posterior / np.sum(posterior)

At the end of a step, the posterior of a step is the prior of next step. 

\begin{equation}
    Posterior_{i} = Prior_{i+1}
\end{equation}

```python
def bernoulli_posterior(n_params=100, n_sample=100, true_p=.8, prior_p=.5, n_prior=100):
    params = np.linspace(0, 1, n_params)
    
    #elimizdeki ornek verilere gore para hileli bir sekilde atiliyor. cunku p degeri 0.8
    sample = np.random.binomial(n=1, p=true_p, size=n_sample)
    
    #likelihood fonksiyonunun bernoulli dagilimi ile hesaplanmasi
    likelihood = np.array([np.product(st.bernoulli.pmf(sample, p)) for p in params])
    #likelihood = likelihood / np.sum(likelihood)
    
    #biz deneyleri yapmadan once gercek p degeri bilgisine(true_p=0.8) sahip omadan paranin hilesiz oldugunu(prior_p=0.5) soyluyoruz.
    prior_sample = np.random.binomial(n=1, p=prior_p, size=n_prior)
    prior = np.array([np.product(st.bernoulli.pmf(prior_sample, p)) for p in params])
    prior = prior / np.sum(prior)
    
    #posterior degeri  prior ve likelihood degerinin carpimina esit 
    posterior = [prior[i] * likelihood[i] for i in range(prior.shape[0])]
    posterior = posterior / np.sum(posterior)
    
    #grafik 
    fig, axes = plt.subplots(3, 1, sharex=True, figsize=(8,8))
    axes[0].plot(params, likelihood)
    axes[0].set_title("Sampling Distribution")
    
    axes[1].plot(params, prior)
    axes[1].set_title("Prior Distribution")
    
    axes[2].plot(params, posterior)
    axes[2].set_title("Posterior Distribution")
    sns.despine()
    plt.tight_layout()
     
    return posterior
```

```python
example_dist = bernoulli_posterior()
```

So, this was first step. At now, we shall use our posterior distribution of first step for prior distribution in second stage. First posterior is about 0.7. This value will be prior for second step. 

```python
second_dist = bernoulli_posterior(prior_p=0.7)
```

```python
dist_with_more_data = bernoulli_posterior(n_sample=1000)
```

The distribution of Likelihood function is narrow towards peak at now. More data decreases spread of a distribution. Also, we see that posterior distribution is closer to likelihood(sampling distributions).
