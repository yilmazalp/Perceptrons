### Logistic Regression

Logistic regression method is a nonlinear regression method which studies the association between categorical dependent variable and independent variables. 

In logistic regression, mathematical model of set of explanatory variable is used for predicting ${logit}$ transformation of the dependent variable. 

Assume that 0 and 1 are two outcomes of binary variable. In general, 0 symbolizes negative responses, and 1 symbolizes positive responses. The mean of this variable is proportion of positive responses. If p is proportion of variables have positive response as outcome, then 1-p is probability of variables have negative outcomes. The ratio of these proportions, p/(1-p), is called \emph{odds} and logarithm of odds is called ${logit}$.

\begin{equation}
    \frac{p}{1-p} : odds
\end{equation}

\begin{align}
    {l} = logit(p) = \ln{\frac{p}{1-p}}
\end{align}

Logistic transformation is inverse of the logit transformation. 

\begin{equation}
    p = {logistic({l})} = \frac{e^{l}}{e^{l}+1} 
\end{equation}

In logistic regression, Y has two unique dependent variable is regressed with n independent variable as $X_{1},X_{2},...,X_{n}  \quad n \in \mathbb{N}$


\begin{align*}
     X = ({X_1},{X_2},...,{X_n})
\end{align*}

\begin{align*}
    \beta = \begin{bmatrix}\\\beta_1\\ \beta_2 \\ .. \\ .. \\ \beta_n\end{bmatrix}
\end{align*}

So, regression model is expressed with the equation is:

\begin{equation}
    \ln{\frac{p}{1-p}} = \alpha + \beta_1{X_1}+\beta_2{X_2}+...+\beta_n{X_n}
\end{equation}

\begin{equation}
    p=P(Y|X)= \frac{e^{\alpha+X\beta}}{1+e^{\alpha+X\beta}}
\end{equation}

#### Cost Function

Cost function is used to reduce difference between target value and output value to minimum. Mean squared error is used to obtain this function in linear regression. However, cost function cannot be used for logistic regression as is the case with linear regression because the function is used in logistic regression is not linear. Cost function has many local optima value since it is not convex function and gradient descent function may not find to global minimum value.

Let $sigmoid(x)=h_\theta(x)$, cost function is computed as

\begin{equation}
    cost(h_{\theta}(x), y)=-ylog(h_{\theta}(x))-(1-y)log(1-h_{\theta}(x))
\end{equation}

#### Gradient Descent

Aim of the model is minimizing of cost function. So, gradient descent algorithm can be used for this process.

\begin{equation}
    \theta_j := \theta_j-\alpha\frac{\partial}{\partial \theta_j}J(\theta) \quad \text{  } j=1,2,..,m
\end{equation}

In every step, partial derivative of cost function is multiplied with learning rate, this result is extracted from old cost function value, and new cost function value is found. This process is repeated through all input data values which means from 1 to m which is number of data.

In stochastic gradient descent, cost gradient of only one example is computed at each iteration instead of sum of the cost gradient of all examples. 


\begin{algorithm}
    \caption{Stochastic Gradient Descent}
    \label{stochastic}
    \begin{algorithmic}
        \For{\texttt{i=1 to m}}
        \State \texttt{$\theta_j := \theta_j-\alpha\frac{1}{m}(h_\theta(x_i)-y_i)x_j^{(i)}$}
        \EndFor
    \end{algorithmic}
\end{algorithm}

In batch gradient descent, sum of the cost gradient of all examples is computed. This process is repeated until convergence for j=1 to m which is number of data. 

\begin{algorithm}
    \caption{Batch Gradient Descent}
    \label{batch}
    \begin{algorithmic}
        \While{\texttt{until convergence}}
        \State \texttt{$\theta_j := \theta_j-\alpha\frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i)x_j^{(i)}$}
        \EndWhile
    \end{algorithmic}
\end{algorithm}

### Perceptron 

Perceptron is basis form of Artificial Neural Networks. It consist of four important parts such as input, weights, summation function, and activation function. 

Aim of perceptron is determine that input values belongs to which classes. Input and weights values are multiplied, result of multiplication is applied to activation function. At the end, input values are classified according to output value. 

Since single layer pairs have not priori knowledge, initial weight  values are assigned as randomly. Perceptron summed weighted input values and if summation is greater than threshold value, then output is +1. If summation is smaller than threshold value, then output is 0 or -1. These output values change based on activation function. 


\begin{align}
    w_0x_0 + w_1x_1 + \cdots + w_nx_n + b > 0 & \Longrightarrow \text output : 1 \\
    w_0x_0 + w_1x_1 + \cdots + w_nx_n + b < 0 & \Longrightarrow \text output : 0
\end{align}

There are different functions which is used as activation function. Most used activation functions are sigmoid function, sign function, and hyperbolic function. 

\begin{algorithm}
    \caption{Perceptron Algorithm}
    \label{perceptron}
    \begin{algorithmic}
        \While{\texttt{until convergence}}
        \State \textbf{Set perceptron}
        \State \textbf{Predict output}
        \State \textbf{Compute error}
        \State \texttt{error = desired output - predicted output}
        \State \textbf{Set all weights according to error:} 
        \State \texttt{$\Delta$weight = error x input}   
        \State \texttt{new weight = weight + $\Delta$weight x learning rate}
        \EndWhile
    \end{algorithmic}
\end{algorithm}

In the algorithm, learning rate is considered for regularization. Learning rate must be assigned as properly to obtain optimized weight values. Assigned astronomic value for learning rate provides that reaching to solution quickly but this circumstance brings about sharp changes in weight values. This avoids finding optimal values. Appointed low value for learning rate leads to spending too many time and power to obtain optimized weight values.

### Bayesian Statistics

Bayesian statistics is a mathematical procedure based on Bayesian approach to probabilities.

#### Conditional Probability

Firstly, it will be better understand conditional probabilities before Bayesian statistics. Consider two events A and B. In conditional probability, the probability of an event A given B equals to probability of B and A divided by probability of B. 

\begin{equation}
    P(A|B) = \frac{P(B|A)}{P(B)}
\end{equation}

or

\begin{equation}
    P(B|A) = \frac{P(A|B)}{P(A)}
\end{equation}

We shall try to understand this with an example:  

Assume that out of all the 4 championship races between Niki Lauda and James hunt, Niki won 3 times while James won only 1 time. 
The question is that who be winner on next race? 

It can be considered that Niki Lauda will win next race in nonconditional situations. However, if the event of raining is introduced to this race and winners are different on rainy days. So, suppose that it rained once when James won and once when Niki won and it is definite that it will rain on the next date. Now, who will be winner on next race? 

Let B be the event of winning of James Hunt. 

\begin{equation}
    P(B) = \frac{1}{4}
\end{equation}

Assume that A is the event of raining. 

\begin{equation}
    P(A) = \frac{1}{2}
\end{equation}

Two days of four days are rainy. 

\begin{equation}
    P(A|B) = 1
\end{equation}

Because it was raining when James Hunt won the race. So, the condition is that Hunt won the race. 

\begin{equation}
    P(A|B) = \frac{P(A\cap B)}{P(B)}
\end{equation}

\begin{equation}
    P(B) = \frac{1}{4}, \quad P(A|B) = 1 
\end{equation}

So, 

\begin{equation}
    P(A\cap B) = \frac{1}{4}
\end{equation}

What is the probability of winning of James Hunt when the wheather is rainy? It means that what is $P(B|A)$? 

\begin{equation}
    P(B|A) = \frac{P(A\cap B)}{P(A)}
\end{equation}

\begin{equation}
    P(B|A) = \frac{\frac{1}{4}}{\frac{1}{2}} = \frac{1}{2}
\end{equation}

#### Bayesian Inference 

The most important part of Bayesian Inference is setting of models and parameters. Models are mathematical expressions of observed events. Parameters are factors which effect on observed data. For instance, in tossing a coin, fairness of a coin can be described parameter of a coin. 

Let the parameter is represented with $w$ which is fairness of a coin and suppose that $D$ is outcome of events. 

In Bayes theorem, 

\begin{equation}
    P(\theta|D) = \frac{P(D|\theta)}{P(D)}
\end{equation}


Multiplying likelihhod function and prior belief gives that posterior  belief. 

\begin{equation}
    P(\theta|D) = P(D|\theta)\text{ * }P(\theta)
\end{equation}

$P(\theta)$: $Prior$, the strength of belief on fairness of a coin before toss a coin. Coin can take any value between 0 and 1.  

$P(D)$: $Evidence$ , is a normalization constant. It guarantess that posterior distribution is valid probability distribution and it is integrated to 1. So, 

\begin{equation}
    P(D) = \int_{}^{} P(D|\theta)P(\theta)d\theta
\end{equation}

$P(D|\theta)$: $Likelihood function$

It is likelihood of obtained given distribution for w parameter. 

$P(\theta|D)$ : $Posterior$

The belief of parameters $after$ obtain result of events.

#### Bernoulli Likelihood Function

$P(y=1|\theta) = \theta^y$ (if coin is fair θ=0.5, probability of observing heads (y=1) is 0.5)

$P(y=0|\theta) = (1-\theta)^{1-y}$ (if coin is fair θ=0.5, probability of observing tails (y=1) is 0.5)

$$P(y|\theta) = \theta^y(1-\theta)^{1-y}$$

If we consider more than 1 toss events, the likelihood function is

$$P(y_{1}, y_{2}, ... , y_{n}|\theta) = {\overset{n}{\underset{i=0}{{\displaystyle\prod}}}P(y_{i}|\theta)}$$


$$ = {\overset{n}{\underset{i=0}{{\displaystyle\prod}}}\theta^{y_{i}}({1-\theta})^{1-y_{i}}}$$

if we are interested in the probability of number of heads z turning up in N number of flips then the probability is given by:

\begin{equation}
    P(z, N|\theta) = \theta^z(1-\theta)^{N-z}
\end{equation}



#### Prior Belief Distribution

This distribution is used to represent the strength of beliefs about parameters based on previous experiences. `Beta distribution` is a mathematical function which is used to represent prior beliefs. Probability of density function of beta distribution is

\begin{equation}
    \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)}
\end{equation}


$B(\alpha, \beta$) is there just to ensure that the total probability density function upon integration evaluates to 1. $\alpha$ is number of heads and $\beta$ is number of tails. 

\begin{equation}
    \mu = \frac{\alpha}{\alpha + \beta}
\end{equation}

\begin{equation}
    \sigma = \sqrt{\frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}}
\end{equation}

#### Posterior Belief Distribution


\begin{equation}
    P(z, N|\theta) = \theta^z(1-\theta)^{N-z}
\end{equation}

\begin{equation}
    P(\theta|z, N) = \frac{P(z, N|\theta)P(\theta)}{P(z, N)}
\end{equation}

\begin{equation}
    = \theta^z(1-\theta)^{N-z} \frac{\theta^{\alpha-1}(1-\theta)^{\beta-1}}{B(\alpha, \beta)P(z,N)}
\end{equation}

\begin{equation}
    = \frac{\theta^{z+\alpha-1}(1-\theta)^{N-z+\beta-1}}{B(z+\alpha, N-z+\beta)}
\end{equation}

\begin{equation}
    = P(\theta|z+\alpha, N-z+\beta)
\end{equation}


If we know mean value, standart deviation, obtain number of heads in N flips, then we can update belief about parameter of model, $\theta$.

```python
#parametrelerin atanmasi
alpha = 6.0
beta = 0.8
sigma = 1
```

```python
#data yaratma ve plot edilmesi 

x = 10*np.random.rand(100) #1'den 100'e kadar rastgele degerlere sahip 100 tane x degeri uretildi
y = alpha + beta*x
y = np.random.normal(y, scale=sigma) #y'nin normalize edilmesi. edilmediginde bir dagilim gostermiyor. 

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot of x and y')
```

```python
model = """
data {
    int<lower=0> N;
    vector[N] x;
    vector[N] y;
}
parameters {
    real alpha;
    real beta;
    real<lower=0> sigma;
}
model {
    y ~ normal(alpha + beta * x, sigma);
}
"""
```

```python
data = {'N': len(x), 'x': x, 'y': y}
```

```python
#modeli derleme(compile)
sm = pystan.StanModel(model_code = model)

#nodeli egitmek(train) ve orneklemleri yaratmak
fit = sm.sampling(data=data, chains=4, iter=500, warmup=250, thin=3, seed=101)
```

##### iter
number of the samples(vectors) in each Markov chain

##### chains
number of the chains from samples which is combined to form the posterior distribution. It is beneficial that the chains which will initialize at different locations in parameter space are more although more chains is costly for fitting. 

##### warmup
number of samples which will discards from the beginning of each Markov chain. When Markov chain had a chance to reach equilibrium, early samples will be drawn. So, samples on the beginning of each Markov chain are discarded. 

##### thin 
the interval for retaining the samples in sampling. If `thin` is equal to 3, then every third sample is retained and rest is discarded. This is important to temper the effect of correlation between successive samples. If `thin` is equal to 1, then every sample is retained. 

##### seed
it is specified to allow for reproducibility.

```python
def Markov_chain(initial_state, transition_matrix, chain_number):
    situation = initial_state
    
    for chain in range(chain_number):
        situation = situation*transition_matrix
        print(situation)
```

### Plotting Posteriors

```python
def plot_trace(param, param_name='parameter'):
    #trace ve bir parametrelerin posterior dagilimlarini gorsellestirme 
    
    # Summary statistics
    mean = np.mean(param)
    median = np.median(param)
    cred_min, cred_max = np.percentile(param, 2.5), np.percentile(param, 97.5)
  
    # Plotting
    plt.subplot(2,1,1)
    plt.plot(param)
    plt.xlabel('samples')
    plt.ylabel(param_name)
    plt.axhline(mean, color='r', lw=2, linestyle='--')
    plt.axhline(median, color='c', lw=2, linestyle='--')
    plt.axhline(cred_min, linestyle=':', color='k', alpha=0.2)
    plt.axhline(cred_max, linestyle=':', color='k', alpha=0.2)
    plt.title('Trace and Posterior Distribution for {}'.format(param_name))

    plt.subplot(2,1,2)
    plt.hist(param, 30, density=True); 
    sns.kdeplot(param, shade=True)
    plt.xlabel(param_name)
    plt.ylabel('density')
    plt.axvline(mean, color='r', lw=2, linestyle='--',label='mean')
    plt.axvline(median, color='c', lw=2, linestyle='--',label='median')
    plt.axvline(cred_min, linestyle=':', color='k', alpha=0.2, label='95% CI')
    plt.axvline(cred_max, linestyle=':', color='k', alpha=0.2)
  
    plt.gcf().tight_layout()
    plt.legend()
```

```python
plot_trace(alpha, 'alpha')
```

```python
plot_trace(beta, 'beta')
```
